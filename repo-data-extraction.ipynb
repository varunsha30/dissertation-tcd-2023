```python
import pydriller
import pandas as pd
import json
from pydriller import Repository
from datetime import datetime as dt
from datetime import timedelta
```

```python
repo_df = pd.read_json('data/link-data.json')
```

```python
repo_df
```

```python
def extract_repository_info(repo_url):
    repository_info = {}
    
    # Extract information from the repository
    repo = Repository(repo_url)
    commits = list(repo.traverse_commits())
    
    repository_info['url'] = repo_url
    repository_info['total_commits'] = len(commits)
    repository_info['commits'] = [commit.hash for commit in commits]
    
    authors = set()
    for commit in commits:
        authors.add(commit.author.name)
    repository_info['total_authors'] = len(authors)
    
    repository_info['first_commit_date'] = commits[-1].committer_date + timedelta(seconds=commits[-1].committer_timezone)
    repository_info['last_commit_date'] = commits[0].committer_date + timedelta(seconds=commits[0].committer_timezone)
    
    return repository_info
```

```python
response = extract_repository_info(repo_url=repo_df.github_link[0])
```
